```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
import numpy as np
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
sns.displot(tips_df['size'])
```

```python
sns.displot(tips_df['tip'])
```

```python
sns.displot(tips_df['total_bill'])
```

```python
sns.displot(tips_df['total_bill'],bins=10)
```

```python
sns.displot(tips_df['total_bill'],bins=100) 
```

```python
sns.displot(tips_df['total_bill'],kde=True) 
```

```python
sns.displot(tips_df['total_bill'],rug=True)
```

```python
sns.set(style='darkgrid')
sns.displot(tips_df['total_bill'],label="Total Bill")

plt.title('Histogram of total_bill',fontsize=15)
legend=()
```

```python
# For calculation of bins
tips_df.total_bill.sort_values()
```

```python
bins = np.arange(0,58,5)
bins
```

```python
plt.figure(figsize=(13,7))

sns.set(style='darkgrid')

sns.distplot(tips_df['total_bill'],bins=bins)
plt.xticks(bins)

plt.title("Histogram of Total Bill")
```

```python
plt.figure(figsize=(13,7))

sns.set(style='darkgrid')

sns.distplot(tips_df['total_bill'],bins=bins,hist=False)
plt.xticks(bins)

plt.title("Histogram of Total Bill")
```

```python
plt.figure(figsize=(13,7))

sns.set(style='darkgrid')

sns.distplot(tips_df['total_bill'],bins=bins,hist=False,fit=norm)
plt.xticks(bins)

plt.title("Histogram of Total Bill")
```

```python
plt.figure(figsize=(13,7))

sns.set(style='darkgrid')

sns.distplot(tips_df['total_bill'],bins=bins,hist=False,fit=norm,kde=False)
plt.xticks(bins)

plt.title("Histogram of Total Bill")
```

```python
plt.figure(figsize=(16,9))
sns.set()
 
# hist keyword argument to change hist format
sns.distplot(tips_df["total_bill"],
            hist_kws = {'color':'#DC143C', 'edgecolor':'#aaff00',
                       'linewidth':5, 'linestyle':'--', 'alpha':0.9}) # hist keyword parameter to change hist format
```

```python
plt.figure(figsize=(16,9))
sns.set()
 
# hist, kde and rug keyword  argument to change hist format
sns.distplot(tips_df["total_bill"],
            hist_kws = {'color':'#DC143C', 'edgecolor':'#aaff00',
                       'linewidth':5, 'linestyle':'--', 'alpha':0.9},
             
            kde_kws = {'color':'#8e00ce', 
                       'linewidth':8, 'linestyle':'--', 'alpha':0.9},
            rug = True,
            rug_kws = {'color':'#0426d0',
                       'linewidth':3, 'linestyle':'--', 'alpha':0.9})
```

```python
#Plot histogram in best format
plt.figure(figsize=(16,9))
sns.set()
 
bins = [1,5,10,15,20,25,30,35,40,45,50,55]
sns.distplot(tips_df["total_bill"],bins=bins,
            hist_kws = {'color':'#DC143C', 'edgecolor':'#aaff00',
                       'linewidth':5, 'linestyle':'--', 'alpha':0.9},
             
            kde=False,
            fit = norm,
            fit_kws = {'color':'#8e00ce', 
                       'linewidth':12, 'linestyle':'--', 'alpha':0.4},
            rug = True,
            rug_kws = {'color':'#0426d0','linewidth':3, 'linestyle':'--', 'alpha':0.9},
            label = "TB")
 
plt.xticks(bins)
plt.title("Histogram of Restorant Total Bill", fontsize = 20)
plt.xlabel("Total Bill", fontsize = 15)
plt.legend()
plt.show()
```

```python
# Plot multiple seaborn histogram in single graph
plt.figure(figsize=(16,9))
sns.distplot(tips_df["total_bill"], bins=9, label="total_bil")
sns.distplot(tips_df["tip"], bins=9, label="tip")
sns.distplot(tips_df["size"], bins=9, label = "size")
 
plt.legend()
```
