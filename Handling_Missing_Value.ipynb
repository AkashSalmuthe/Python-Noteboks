```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Dataset

df = pd.read_csv('/content/drive/MyDrive/AI Production/5. Data Preprocessing and Feature Engineering/house_train.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(25,20))

sns.heatmap(df.isnull())
```

```python
# Finding total % of missing value in each row

null_var = df.isnull().sum()/df.shape[0] *100
null_var
```

```python
# Drop columns with values 17% and above

drop_col = null_var[null_var>17].keys()
drop_col
```

```python
df2_drop_col = df.drop(columns=drop_col)
df2_drop_col.shape
```

```python
sns.heatmap(df2_drop_col.isnull())
```

```python
# Drop rows with na values
df3_drop_rows = df2_drop_col.dropna()
df3_drop_rows.shape
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(df3_drop_rows.isnull())
```

```python
# Get Total missing value count in dataframe
df3_drop_rows.isnull().sum().sum()
```
