```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
ar_2d = np.linspace(1,5,12).reshape(4,3)
ar_2d
```

```python
sns.heatmap(ar_2d)
```

```python
globalWarming_df = pd.read_csv("/content/drive/MyDrive/AI Production/4. Seaborn/Who_is_responsible_for_global_warming.csv")
globalWarming_df
```

```python
globalWarming_df  = globalWarming_df.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code'], axis=1).set_index('Country Name')
globalWarming_df
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df,vmin=0,vmax=21)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df,cmap="coolwarm")
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df,cmap="gnuplot2_r",annot=True)
```

```python
ar_2d
```

```python
annot_arr = np.array([['a00','a01','a02'],
                      ['a10','a11','a12'],
                      ['a20','a21','a22'],
                      ['a30','a31','a32']], dtype = str)
 
sns.heatmap(ar_2d, annot = annot_arr, fmt="s") # s -string, d - decimal
```

```python
# annot_kws parameter
 
plt.figure(figsize=(16,9))
 
annot_kws={'fontsize':10, # 
           'fontstyle':'italic',  
           'color':"k",
           'alpha':0.6, 
           'rotation':"horizontal",
           'verticalalignment':'center',
           'backgroundcolor':'w'}
 
sns.heatmap(globalWarming_df, annot = True,cmap='gnuplot2_r', annot_kws= annot_kws)
```

```python
# annot_kws parameter
 
plt.figure(figsize=(16,9))

sns.heatmap(globalWarming_df, annot = True,cmap='gnuplot2_r', linewidth=3)
```

```python
# annot_kws parameter
 
plt.figure(figsize=(16,9))

sns.heatmap(globalWarming_df, annot = True,cmap='gnuplot2_r', linewidth=3,linecolor='k')
```

```python
plt.figure(figsize=(16,9))

sns.heatmap(globalWarming_df, annot = True,cmap='gnuplot2_r', linewidth=3,linecolor='#58FAF4',cbar=False)
```

```python
plt.figure(figsize=(16,9))

cbar_kws={"orientation":"horizontal", 
            "shrink":1,
            'extend':'min', 
            'extendfrac':0.1, 
            "ticks":np.arange(0,22), 
            "drawedges":True,}

sns.heatmap(globalWarming_df,cmap="gnuplot2_r",annot=True,cbar_kws=cbar_kws)
```

```python
# change y-axis labels using yticklabels parameter
 
plt.figure(figsize=(16,9))
 
country_code = ['USA', 'GBR', 'IND', 'CHN', 'RUS', 'AUS', 'FRA', 'DEU', 'CAN', 'BRA', 'ARG', 'PAK', 'NPL', 'BGD', 'JPN']
 
sns.heatmap(globalWarming_df, yticklabels = country_code)
```

```python
# set seaborn heatmap title, x-axis, y-axis label and font size
 
plt.figure(figsize=(16,9))
 
ax = sns.heatmap(globalWarming_df)
 
ax.set(title="Heatmap",
      xlabel="Years",
      ylabel="Country Name",)
 
sns.set(font_scale=2) # set fontsize 2
```

# Correlation

```python
globalWarming_df.corr()
```

```python
plt.figure(figsize=(16,9))

cor = globalWarming_df.corr()

ax = sns.heatmap(cor,annot=True,linewidth=3)
ax.tick_params(size=10,color='w,',labelsize=10)

plt.title("Heatmap of Global Wrming",fontsize=20)
```

```python
from sklearn.datasets import load_breast_cancer
cancer_df = load_breast_cancer()
cancer_df
```

```python
cancer_df = pd.DataFrame(np.c_[cancer_df['data'],cancer_df['target']],
                         columns=[np.append(cancer_df['feature_names'],['target'])])
cancer_df
```

```python
plt.figure(figsize=(25,25))

cor = cancer_df.corr()

ax = sns.heatmap(cor,annot=True,linewidth=3)
ax.tick_params(size=10,color='w,',labelsize=10)

plt.title("Heatmap of Breast Cancer ",fontsize=20)
plt.show()
```
