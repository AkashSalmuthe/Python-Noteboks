# **One Hot**

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oh_enc = OneHotEncoder(sparse=False,drop='First')
```

```python
oh_enc_arr = oh_enc.fit_transform(tips_df[['sex','smoker','day','time']])
oh_enc_arr
```

```python
dummy_df.keys()
```

```python
oh_enc_df = pd.DataFrame(oh_enc_arr,columns=['sex_Male', 'sex_Female', 'smoker_Yes',
       'smoker_No', 'day_Thur', 'day_Fri', 'day_Sat', 'day_Sun', 'time_Lunch',
       'time_Dinner']) 
oh_enc_df
```
