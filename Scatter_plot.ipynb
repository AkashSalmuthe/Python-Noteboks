```python
df = pd.read_csv('/content/drive/MyDrive/AI Production/Matplotlip/googleplaystore.csv',nrows=1000)
```

```python
x,y = df['Rating'],df['Reviews']
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(13,7))
style.use('ggplot')

plt.scatter(x,y)

plt.title('Play Store App Scatter Plot')
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(10,5))
style.use('ggplot')

plt.scatter(x,y,c='r',marker = '*')

plt.title('Play Store App Scatter Plot')
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(10,5))
style.use('ggplot')

plt.scatter(x,y,c='r',marker = '*',s=100) # Increase the size of marker

plt.title('Play Store App Scatter Plot')
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(10,5))
style.use('ggplot')

plt.scatter(x,y,c='r',marker = '*',s=100,alpha=0.5)

plt.title('Play Store App Scatter Plot')
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(10,5))
style.use('ggplot')

plt.scatter(x,y,c='r',marker = '*',s=100,linewidths=5)

plt.title('Play Store App Scatter Plot')
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(15,9))
style.use('ggplot')

plt.scatter(x,y,c='r',marker = '*',s=200,linewidths=0.5,edgecolor='k')

plt.title('Play Store App Scatter Plot')
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(15,9))
style.use('ggplot')

plt.scatter(x,y,c='r',marker = '*',s=200,linewidths=0.5,edgecolor='k')
plt.scatter(x,df['Installs'],c='g',marker = 'o',s=200,linewidths=0.5,edgecolor='b')

plt.title('Play Store App Scatter Plot')
plt.xlabel("Rating")
plt.ylabel("Reviews and Installs")
```
