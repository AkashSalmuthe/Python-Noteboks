```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_student_age = np.random.randint(18,45,(100)) 
py_student_age = np.random.randint(15,40,(100))
```

```python
print('ml_student_age: \n',ml_student_age)
print()
print('py_student_age: \n',py_student_age)
```

```python
plt.hist(ml_student_age)
plt.show()
```

```python
plt.hist(ml_student_age)

plt.title("ML Students age Histogram")
plt.xlabel('Students age categoriey')
plt.ylabel('Number of Students age')

plt.show() 
```

```python
# Defining bins parameter
bins=[15,20,25,30,35,40,45] # Why start with 15 coz Its the lowest range of student 45 is highest age


plt.hist(ml_student_age,bins)

plt.title("ML Students age Histogram")
plt.xlabel('Students age categoriey')
plt.ylabel('Number of Students age')


plt.show() 
```

```python
plt.hist(ml_student_age,bins,rwidth=0.7)
# rwidth by default is 1

plt.title("ML Students age Histogram")
plt.xlabel('Students age categoriey')
plt.ylabel('Number of Students age')


plt.show() 
```

```python
plt.hist(ml_student_age,bins,rwidth=0.7,histtype='step')
# rwidth by default is 1

plt.title("ML Students age Histogram")
plt.xlabel('Students age categoriey')
plt.ylabel('Number of Students age')


plt.show() 
```

```python
plt.hist(ml_student_age,bins,rwidth=0.7,histtype='bar',color='r')

plt.title("ML Students age Histogram")
plt.xlabel('Students age categoriey')
plt.ylabel('Number of Students age')


plt.show() 
```

```python
plt.hist(ml_student_age,bins,rwidth=0.7,histtype='bar',color='y',label="ML Students")

plt.title("ML Students age Histogram")
plt.xlabel('Students age categoriey')
plt.ylabel('Number of Students age')

plt.legend()


plt.show() 
```

```python
plt.hist(ml_student_age,bins,rwidth=0.7,histtype='bar',color='y',label="ML Students")
plt.hist(py_student_age,bins,rwidth=0.7,histtype='bar',color='b',label="Python Students")

plt.title("ML Students age Histogram")
plt.xlabel('Students age categoriey')
plt.ylabel('Number of Students age')

plt.legend()


plt.show() 
```

```python
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.7,histtype='bar',color=['y','b'],label=["ML Students","Python Students"])

# plt.hist(py_student_age,bins,rwidth=0.7,histtype='bar',color='b',label="Python Students")

plt.title("ML Students age Histogram")
plt.xlabel('Students age categoriey')
plt.ylabel('Number of Students age')

plt.legend()


plt.show() 
```

```python
plt.figure(figsize=(13,7))

plt.hist([ml_student_age,py_student_age],bins,rwidth=0.7,histtype='bar',color=['y','b'],label=["ML Students","Python Students"])

plt.title("ML Students age Histogram")
plt.xlabel('Students age categoriey')
plt.ylabel('Number of Students age')

plt.legend()
plt.show() 
```
