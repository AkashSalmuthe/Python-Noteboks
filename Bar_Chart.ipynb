```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes = ["Python", "R","AI","ML",'DS']
class1_students = [30,10,20,25,10]
class2_students = [40,5,20,20,10]
class3_students = [35,5,30,15,15]
```

```python
plt.bar(classes,class1_students)
```

```python
plt.barh(classes,class1_students)
```

```python
plt.bar(classes,class1_students,width=0.5)
```

```python
plt.bar(classes,class1_students,align="edge")
```

```python
plt.bar(classes,class1_students,color='y')
```

```python
plt.bar(classes,class1_students,edgecolor='r')
```

```python
plt.bar(classes,class1_students,edgecolor='r',alpha=0.5)
# alpha for bulrring effect
```

```python
plt.bar(classes,class1_students,edgecolor='r',linewidth=2,linestyle="--")
```

```python
plt.bar(classes,class1_students,edgecolor='r',linewidth=2,linestyle="--",label='Class1_students')
```

```python
plt.bar(classes,class1_students,edgecolor='r',linewidth=2,linestyle="--",visible=False)
```

```python
plt.bar(classes,class1_students,color='k',edgecolor='r',linewidth=2,linestyle="--",visible=True)
```

```python
style.use('ggplot')
plt.bar(classes,class1_students,color='k',edgecolor='r',linewidth=2,linestyle="--")
```

```python
plt.figure(figsize=(13,7))
plt.bar(classes,class1_students,color='k',edgecolor='r',linewidth=2,linestyle="--",visible=True)
```

```python
style.use('ggplot')
plt.bar(classes,class1_students,color='k',edgecolor='r',linewidth=2,linestyle="--")

plt.title("Bar chart of classes")
plt.xlabel("Classes")
plt.ylabel("Students")
```

```python
style.use('ggplot')
plt.bar(classes,class1_students,color='b',edgecolor='r',linewidth=2,linestyle="--",label="Class_1")
plt.bar(classes,class2_students,color='y',edgecolor='r',linewidth=2,linestyle="--",label='Class_2')

plt.legend()

plt.title("Bar chart of classes")
plt.xlabel("Classes")
plt.ylabel("Students")
```

```python
style.use('ggplot')

classes_index = np.arange(len(classes))

plt.bar(classes_index,class1_students,color='b',edgecolor='r',linewidth=2,linestyle="--",label="Class_1")
plt.bar(classes_index,class2_students,color='y',edgecolor='r',linewidth=2,linestyle="--",label='Class_2')

plt.legend()

plt.title("Bar chart of classes")
plt.xlabel("Classes")
plt.ylabel("Students")
```

```python
style.use('ggplot')

classes_index = np.arange(len(classes))

width=0.4

plt.bar(classes_index,class1_students,width,color='b',edgecolor='r',linewidth=2,linestyle="--",label="Class_1")
plt.bar(classes_index+width,class2_students,width,color='y',edgecolor='r',linewidth=2,linestyle="--",label='Class_2')

plt.legend()

plt.title("Bar chart of classes")
plt.xlabel("Classes")
plt.ylabel("Students")
```

```python
style.use('ggplot')

classes_index = np.arange(len(classes))

width=0.4

plt.bar(classes_index,class1_students,width,color='b',edgecolor='r',linewidth=2,linestyle="--",label="Class_1")
plt.bar(classes_index+width,class2_students,width,color='y',edgecolor='r',linewidth=2,linestyle="--",label='Class_2')

plt.xticks(classes_index+width,classes)

plt.legend()

plt.title("Bar chart of classes")
plt.xlabel("Classes")
plt.ylabel("Students")
```

```python


style.use('ggplot')
plt.figure(figsize=(13,7))

classes_index = np.arange(len(classes))

width=0.3

plt.bar(classes_index,class1_students,width,color='b',edgecolor='r',linewidth=2,linestyle="--",label="Class_1")
plt.bar(classes_index+width,class2_students,width,color='y',edgecolor='r',linewidth=2,linestyle="--",label='Class_2')
plt.bar(classes_index+width+width,class3_students,width,color='g',edgecolor='r',linewidth=2,linestyle="--",label='Class_3')

plt.xticks(classes_index+width,classes)

plt.legend()

plt.title("Bar chart of classes")
plt.xlabel("Classes")
plt.ylabel("Students")
```
